## Pie chart

```python
#import matplotlib.pyplot as plt
#import seaborn as sns
#import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Draw Pie chart using the following information

language = ['Python','R','Java','HTML', 'C++']
popularity = [22.8, 16, 10, 8, 9]

colors = ['#FA8F0A','#F0DA32','#61F527','#18DE9D','#3594FF']
```


<a href="https://www.webucator.com/article/python-color-constants-module/">Python Color Constants </a>

```python
plt.pie(popularity,labels=language, colors=  colors)
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
#Move the values inside the pie chart
plt.pie(popularity,labels=language, colors=  colors, autopct = "%1.1f%%")
plt.title("Popularity of Programming Languages")
plt.show()
```

```python
#Explode the hightes value from the chart

explode = (0.2, 0,0,0,0)
plt.pie(popularity,labels=language, colors=  colors, autopct = "%1.1f%%", explode = explode)
plt.title("Popularity of Programming Languages")
plt.show()
```

## Histrogram

```python
# Use the following data to draw a Histrogram
values = [82,76,24,67,40,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52,65,85]
plt.hist(values)
plt.show()
```

```python
# Add in the above histrogram edgecoloe (red), bins = 10, label, and title

values = [82,76,24,67,40,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52,65,85]
plt.hist(values, bins=10, edgecolor='r', label ="test score data")
plt.title("Histogram for scores")
plt.xlabel("scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# change the orientation of the histrogram to horizontal

values = [82,76,24,67,40,62,75,78,71,32,98,89,78,67,72,82,87,66,56,52,65,85]
plt.hist(values, bins=10, edgecolor='k', label ="test score data", orientation='horizontal')
plt.title("Histogram for scores")
plt.xlabel("scores")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## Line Plot

```python
#Use the following information to draw a line plot. don't forget labels, color and legend
''' year = [1960, 1970, 1980, 1990, 2000, 2010]
pop_china = [667.07, 818.32, 981.24, 1135.20, 1262.60, 1337.70]'''


year = [1960, 1970, 1980, 1990, 2000, 2010]
pop_china = [667.07, 818.32, 981.24, 1135.20, 1262.60, 1337.70]

plt.plot(year,pop_china,label='China',color='y')
plt.xlabel("Country Names")
plt.ylabel("Pop data")
plt.legend()
plt.show()
```

```python
# Use the following data with the above one to draw one line plot including all of them
''' pop_india = [449.48, 553.58, 696.78, 870.13, 1053.10, 1231.00]
pop_usa =  [180.67, 205.05, 227.23, 249.62, 282.162, 309.34]'''

pop_india = [449.48, 553.58, 696.78, 870.13, 1053.10, 1231.00]
pop_usa =  [180.67, 205.05, 227.23, 249.62, 282.162, 309.34]

plt.plot(year,pop_china,label='China',color='y')
plt.plot(year,pop_india,label='India',color='b')
plt.plot(year,pop_usa,label='USA',color='k')
plt.xlabel("Country Names")
plt.ylabel("Pop data")
plt.legend()
plt.show()
```

## BoxPlot

```python
# import pandas and then read the file "cancer.csv"

import pandas as pd
df = pd.read_csv("cancer.csv")

#show an overview of  your tabuleted data


df.head()
```

```python
#draw boxplot of area mean using matplotlib lib.

plt.boxplot(df['area_mean'])
```

```python
# Change the orientation of the above boxplot

plt.boxplot(df['area_mean'], vert=False)
```

```python
#draw box plot of the area means (column) that is grouped by diagnosis (column) using matplotlib lib

df.boxplot(column = 'area_mean',by='diagnosis')
plt.title("")
plt.show()
```

```python
#draw box plot of the area means (column) that is grouped by diagnosis (column) using seaborn lib

sns.boxplot(x='diagnosis',y='area_mean',data=df)
```
