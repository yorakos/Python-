```python
#mport pandas as pd
#import matplotlib.pyplot as plt
#import seaborn as sns
#import numpy as np

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Upload tips data from seaborn dataset
tips = sns.load_dataset("tips")
```

```python
# tips is a DataFram type. show the tabulated data
tips.head()
```

```python
#Show the 5 number summary
tips.describe()
```

## Scatter plot

```python
#Draw Sacatter plot for total bill vs tip

sns.scatterplot(x='total_bill',y='tip',data=tips )
```

```python
# Add hue option and make it equal to sex column to relat gender with the other variables

sns.scatterplot(x='total_bill',y='tip',data=tips, hue='sex' )
```

```python
#Fit your data with regression model using seaborn (sns.lmplot)

# The following plot help to estimate the data using regression model (You will learn about regression later )

sns.lmplot(x='total_bill',y='tip', data=tips, hue='sex')
```

```python
#Change the point style of your dat points by using markers option

sns.lmplot(x='total_bill',y='tip', data=tips, hue='sex', markers=['o','x'])
```

```python
#Draw your above graph using two separet facets (graphs)
sns.lmplot(x='total_bill',y='tip', data=tips, col='sex')
```

## Scatter Matrix and Heat maps

```python
#read the file caer.csv using panadas

import pandas as pd

df=pd.read_csv('cars.csv')

#show the head of your tabulated data

df.head()
```

```python
#Draw scatter martix using seaborn

sns.pairplot(df)

# In the diagonal you will see the distrbution of each column
```

```python
#Some time you need to drop your qualtitvtie data in your dataset. It is located in the first column

#We need to drop the first column because it is string and the heat map accept only numerical values in your data
df = df.drop(columns=df.columns[0])

df.head()
```

```python
#Another way to see the correlations for all variables is by creating the heat map

# First we need to find the values of the correlations
corr = df.corr()

#You need to get a suitable figure size to be readable
plt.figure(figsize=(10,10))

#Draw the heat map
sns.heatmap(corr, annot=True) 
```

```python
#Change the color map, you need to use cmap options

corr = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, cmap='BrBG')
```
